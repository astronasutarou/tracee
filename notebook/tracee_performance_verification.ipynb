```python
import numpy as np
import tracee
```

```python
basename = 'sample/mockdata_N005F030D200d00_{}.txt'
for n in range(1401,1406):
    filename = basename.format(n)
    data = np.loadtxt(filename)
    print(f'# filename: {filename}')
    print('## timeit for generate_edge:')
    %timeit edge = tracee.generate_edge(data, 3, 200)
    edge = tracee.generate_edge(data, 3, 200)
    print('## {} edges found'.format(len(edge)))
    print('## timeit for extract:')
    %timeit tracklet = tracee.extract(data, n_neighbor=3)
    tracklet = tracee.extract(data, n_neighbor=3)
    print('## {} tracklets found.'.format(len(tracklet)))
    print('')
```
